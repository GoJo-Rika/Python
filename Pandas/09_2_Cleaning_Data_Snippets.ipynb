```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Mit', 'Steve', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Patel', 'Doe', 'Smith', 'Morris', np.nan, np.nan, 'Missing'],
    'email': ['mitpatel@gmail.com', 'stevedoe@email.com', 'chrismorris@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['28', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)

df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis='index', how='all', subset=['last', 'email'])
```

```python
df.isna()
```

```python
df.fillna(0)
```
