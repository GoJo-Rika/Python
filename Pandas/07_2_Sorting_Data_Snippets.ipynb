```python
people = {
    'first': ['Mit', 'Steve', 'John', 'Adam'],
    'last': ['Patel', 'Rogers', 'Doe', 'Smith'],
    'email': ['mitpatel@gmail.com', 'steverogers@email.com', 'johndoe@email.com', 'adamsmith@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.sort_values(by='last', ascending=False)
```

```python
df.sort_values(by=['last', 'first'], ascending=False)
```

```python
df.sort_values(by=['last', 'first'], ascending=[False, True], inplace=True)
```
