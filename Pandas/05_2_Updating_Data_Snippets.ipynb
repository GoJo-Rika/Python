```python
people = {
    "first": ["Mit", 'Rowland', 'Vish'],
    "last": ["Patel", 'Doe', 'Doe'],
    "email": ["mitpatel@gmail.com", 'rowlanddoe@email.com', 'vishdoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.columns
```

```python
df.columns = ['First_Name', 'Last_Name', 'Email']
```

```python
df
```

```python
df.columns = [x.lower() for x in df.columns]
df
```

```python
df.rename(columns={'first_name': 'first', 'last_name': 'last'}, inplace=True)
```
